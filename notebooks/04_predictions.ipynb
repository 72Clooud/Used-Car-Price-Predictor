```python
import pandas as pd
import numpy as np
import joblib

MODEL_PATH = "../models/best_random_forest_model.joblib"

model_pipeline = joblib.load(MODEL_PATH)
```

```python
def predict_car_price(car_data):

    input_df = pd.DataFrame([car_data])
    log_prediction = model_pipeline.predict(input_df)
    real_price = np.expm1(log_prediction)[0]
    
    return round(real_price, 2)
```

```python
my_car = {
    'brand': 'BMW',
    'year': 2021,
    'usedornew': 'Used',
    'transmission': 'Automatic',
    'engine': 5.0,
    'drivetype': 'Front',
    'fueltype': 'Hybrid',
    'fuelconsumption': 4.5,
    'kilometres': 35000,
    'cylindersinengine': 4,
    'bodytype': 'Sedan',
    'doors': 4,
    'seats': 5,
    'state': 'Nsw',
    'colour_ext': 'red',
    'colour_int': 'Black'
}

price_1 = predict_car_price(my_car)
print(f"Predicted price for your car ({my_car['brand']} {my_car['year']}): ${price_1:,.2f}")
```
